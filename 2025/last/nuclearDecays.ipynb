```python
import numpy as np
import matplotlib.pyplot as plt
U = np.random.rand
```

```python
N  = 100
p  = 10**-3
n  = 1000
Δt = 1 
t  = n*Δt

Ns = [N]

for i in range(t):
    Ns += [N]  # tracks the number of nuclear cores that are alive
    ΔN = 0     # track how many will decay in this round

    for _ in range(N):
        if U() < p: ΔN -= 1 
            
    N = N + ΔN
            

    
# print(Ns)
plt.plot(Ns)
```
