```python
# This is known as Box–Muller transform
import numpy as np
import matplotlib.pyplot as plt
```

```python
σ = 1
def Finv(x): return (-2*σ*σ*np.log(x))**0.5   # equivalent for RNG 
U = np.random.rand
N = 10000
Rs = [ Finv(U()) for i in range(N) ] 
φs = [ U()*2*np.pi for i in range(N)]
Rs = np.array(Rs)
φs = np.array(φs)
Xs = Rs*np.cos(φs)
Ys = Rs*np.sin(φs)
```

```python
plt.hist(Xs, bins = np.linspace(-3, 3, 61), ec = 'black')
```

```python
plt.hist(Ys, bins = np.linspace(-3, 3, 61), ec = 'black')
```
