```python
import numpy as np               # for arrays, bins
import matplotlib.pyplot as plt  # for graphics
```

```python
def U(): return np.random.rand()
```

```python
def f(x) : 
    a, b = 0, 5
    return np.exp(x)/(np.exp(b) - np.exp(a))

fmax = f(5)

hits = []
for i in range(1000000):
    x = (5 - 0)*U() + 0
    y = fmax*U()
    if y <= f(x):
        hits += [x]

plt.hist(hits, ec = 'black')
```

```python
# this is not part of the lecture, answering Vangelis' question, I re-reun the foor loop to create the scatter plot 
hits_x = []
hits_y = []

miss_x = []
miss_y = []

for i in range(1000):
    x = (5 - 0)*U() + 0
    y = fmax*U()
    if y <= f(x):
        hits_x += [x]
        hits_y += [y]
    else:
        miss_x += [x]
        miss_y += [y]
        

plt.scatter(hits_x, hits_y, marker = 'x', color = 'yellow')
plt.scatter(miss_x, miss_y, marker = '.', color = 'black')
```
