```python
import numpy as np
import matplotlib.pyplot as plt

# Define physical constants (SI units)
h = 6.62607015e-34      # Planck constant (J·s)
c = 2.99792458e8        # Speed of light in vacuum (m/s)
kB = 1.380649e-23       # Boltzmann constant (J/K)
```

a)

$I(\lambda) = \frac{2 h c^2 \lambda^{-5}}{e^{hc / (\lambda k_B T)} - 1}$

θέτω

$I(\lambda) \equiv \frac{f}{g}$ με $f = f(\lambda) \equiv 2 h c^2 \lambda^{-5}$ και $g = g(x) \equiv e^{x} - 1$ με $x = x(\lambda) \equiv \frac{hc} {k_B T \lambda}$

$\frac{dI}{d\lambda} = \frac{f'g - g'f}{g^2} = 0$

με $f' = \frac{df}{d\lambda} = -10hc^2 \lambda^{-6}$ και $g'(x) = e^{x}\frac{dx}{d\lambda} = - e^{x}\frac{hc}{k_B T \lambda^2}$

εύκολα προκύπτει ότι το $f'g - g'f = 0$ ισοδυναμεί με την εξίσωση $5 e^{-x} + x - 5 = 0$


b) 

γράφω την εξίσωση $5 e^{-x} + x - 5 = 0$ ως $ x  = 5 - 5 e^{-x}$

```python
# fixed point iterations
def g(x): return 5 - 5*np.exp(-x)
x = 1
for i in range(100):
    if abs(x - g(x))<1.e-6: break
    x = g(x)

b = (h*c)/(kB*x)
print('x = %2.6f'%x)
print('b = %2.10f'%b)
print('NIST value for b is 2.897 771 955...  x 10-3 m K')  
print('[1] https://physics.nist.gov/cgi-bin/cuu/Value?bwien')
```
