```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

```python
fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')

U = lambda : np.random.rand()

N = 100000
R = 3

xhits = [] # 3 lists to save coordinates
yhits = [] # of all points inside the sphere
zhits = []

s    = [] # density color map
for _ in range(N):
    x = 2*R*U() - R 
    y = 2*R*U() - R 
    z = 2*R*U() - R 

    if x**2 + y**2 + z**2 < R**2:
        xhits += [x]
        yhits += [y]
        zhits += [z]
        
        s += [x**2 + y**2] # this is just for the color map

    

# Data for three-dimensional scattered points
ax.scatter3D(xhits, yhits, zhits, c = s, marker="x", cmap = 'hot', s =1);

# Set the axis labels
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.savefig('sphere3d.pdf')
```
