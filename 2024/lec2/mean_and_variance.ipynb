```python
xs = [1,2,2,3,3,3,4,4,4,4]
Σx  = 0
Σx2 = 0

for i in range(0, len(xs)):
    Σx  = Σx + xs[i] 
    Σx2 = Σx2 + xs[i]*xs[i] 

N = len(xs)
μ = Σx / N
σ2 = Σx2/N - μ*μ
s2 = (N/(N-1))*σ2 

print('μ = ', μ, 's2 = ', s2, 'σ2 = ', σ2)
```

```python
import matplotlib.pyplot as plt

plt.hist(xs, bins = range(0,7), ec = 'black')
```
