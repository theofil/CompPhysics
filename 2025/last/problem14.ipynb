```python
import numpy as np
U = np.random.rand
```

```python
def f(x, y): return (20/13.)*(x + y)

N   = 100000
n   = 0     # hits
Σf  = 0
Σf2 = 0

for i in range(N):
    x = U()
    y = U()
    
    if y >= x**2:
        n   += 1
        Σf  += f(x, y)          # w = f(x,y)
        Σf2 += f(x, y)*f(x, y)  # w*w
        
μ  = Σf/n
σ2 = Σf2/n - μ**2
s2 = σ2*(n/(n-1))
s  = s2**0.5
V  = 2/3.
I  = V*μ 
δI = V*s/n**0.5

# print(I, δI)
print('I = %2.3f +/- %2.3f'%(I, δI))
```
