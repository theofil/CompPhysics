```python
ln = np.log
exp = np.exp

def f(x): 
    return exp(x)/(exp(5) - 1)

fmax = f(5)

N = 10000
xs = []

for i in range(N):
    x = 5*U()
    y = fmax*U()
    if y < f(x): xs += [x]
        
        
plt.hist(xs, bins = np.arange(0, 5, 0.5), ec='black')    
```
