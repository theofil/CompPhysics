```python
import numpy as np
import matplotlib.pyplot as plt
U = np.random.rand
```

```python
def step():
    step = -1
    if U() > 0.5: step = 1
    return step
```

```python
M = 1000 # number of walkers 
N = 10   # number of steps

xs = []
for i in range(M):
    x = 0
    for i in range(N):
        x += step() 
    xs += [x] 
    
# print(xs)

plt.hist(xs, ec = 'black', bins = np.arange(-12, 13, 1))
```
