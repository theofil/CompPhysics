```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
U = np.random.rand
def step():
    res = 1
    if (U()<0.5): res = -1
    return res
```

```python
# one walker case
N  = 30000
x  = 0
xs = [] 
for i in range(N):
    xs += [x]
    x = x + step()

plt.plot(xs)
```

```python
def walk(N = 3):
    '''returns distance at termination of one walker'''
    x  = 0
    for i in range(N):
        x = x + step()
    return x   
```

```python
walks = [walk(N=64) for i in range(10000)]
plt.hist(walks, ec = 'black', bins = np.linspace(-24,24,49))
```

```python
# verifying that the std of the walks is about 64
print(np.array(walks).std())
```
