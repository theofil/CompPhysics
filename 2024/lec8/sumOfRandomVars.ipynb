```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def U(): return np.random.rand()
```

```python
N = 10000
### generate the random numbers using list comprehension

# x1 is a list of N uniform random vars
x1 =  [U() for i in range(N)] 

# x2 is another list of N uniform random vars
x2 =  [U() for i in range(N)] 

# x is list of the sum of two uniform random vars
x = [U() + U() for i in range(N)] 
```

```python
### alternative code to generate the random numbers using a for loop
# x1 = [] 
# x2 = []
# x  = []
# for i in range(N):
#     x1 += [U()]
#     x2 += [U()]
#     x  += [U() + U()]
```

```python
plt.hist(x1, ec = 'black')
```

```python
plt.hist(x2, ec = 'black')
```

```python
plt.hist(x, ec = 'black', bins = 100)
```

```python
# smart line example showing the sum of 6 uniform random vars
plt.hist([ U() + U() + U() + U() + U() + U() for i in range(N)], bins = 20, ec = 'black')
```

```python
# demonstration of central limit theorem, using the sum of 10 uniform random variables
xbars = []
for j in range(100000):
    x = 0
    for i in range(10):
        x += U() 
    
    xbar = x/10  
    xbars += [xbar]
```

```python
plt.hist(xbars, bins = 100, ec = 'black')
```
