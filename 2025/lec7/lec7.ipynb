```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
U = np.random.rand
```

```python
xs = []
for i in range(100000):
    x = U() + U()  # != 2*U()
#     print(x)
    xs = xs + [x] # xs += [x] 

# plt.figure(figsize=(3,3))

_ = plt.hist(xs, ec = 'black', bins = 100)
```

```python
xs = []
for i in range(10000):
    x = U() + U() + U() 
    xs = xs + [x] # xs += [x] 

plt.figure(figsize=(3,3))

_ = plt.hist(xs, ec = 'black', bins = 50)
```

```python
x10s = []
x5s  = []
for i in range(10000):
    x10 = (U() + U() + U() + U() + U() + U() + U() + U() + U() + U())*0.1
    x5 =  (U() + U() + U() + U() + U())*0.2

    x10s = x10s + [x10] 
    x5s  = x5s  + [x5]

plt.figure(figsize=(3,3))

_ = plt.hist(x5s , ec = 'black', bins =  np.arange(0, 1, 0.01) )
_ = plt.hist(x10s, ec = 'black', bins =  np.arange(0, 1, 0.01) )
```
