```python
import numpy as np
import matplotlib.pyplot as plt
U = np.random.rand
```

```python
xs = []
ys = []


for i in range(1000):
    x = 2*U() - 1
    y = 2*U() - 1
    if x**2 + y**2 <= 1:
        xs += [x]
        ys += [y]

plt.scatter(xs, ys)
# _ = plt.hist2d(xs, ys)    
```
