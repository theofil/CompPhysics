```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def U(): return np.random.rand()
def Finv(x): 
    a, b = 0, 5
    
    return np.log(x*( np.exp(b) - np.exp(a)) + np.exp(a))
```

```python
# I do it with for loop
myXs = []

for i in range(100000):
    u = U()
    myX = Finv(u)
    myXs += [myX]

plt.hist(myXs, ec = 'black')
```

```python
# I do from scracth it with numpy arrays and list comprehension
Us = np.array([U() for _ in range(10000)])
Xs = Finv(Us) 
plt.hist(Xs, ec = 'black', bins = np.linspace(0,5,100))
```
